```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      print("大人です")
    else:
      print("大人ではありません")

human = Human("侍太郎", 20)
human2 = Human("侍花子", 19)
human_list = [human, human2]

for i in human_list:
  i.check_adult()

```
